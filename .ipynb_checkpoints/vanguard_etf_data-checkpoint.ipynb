```python
import yfinance as yf

# vanguard etfs
data = yf.download("VOX VCR VDC VIG VWOB VDE VSGX VCEB", start="1995-01-01", end="2022-12-31")
```

```python
data
```

```python
import pandas as pd

data.to_csv('vanguard_etfs.csv')
```

```python
df = pd.read_csv('vanguard_etfs.csv')
```

```python
df.drop([1, 1])
```

```python
df
```

```python
new_df = df.rename(columns={"Unnamed: 0": "Date"})
```

```python
new_df
```

```python
new_df.to_csv('test.csv', index=False)
```

```python
updated_df = new_df.drop([1, 1])
```

```python
updated_df
```

```python
updated_df.to_csv('vanguard_etfs.csv', index=False)
```

```python
updated_df
```

```python
# sorting by largest Adj Close.1 values
updated_df.groupby(['Date']).mean().sort_values('Adj Close.1', ascending=False)
```

```python
# sorting by smallest Adj Close.1 values
updated_df_final = updated_df.groupby(['Date']).mean().sort_values('Adj Close.1', ascending=True)
```

```python
# saving initial data to excel workbook
data.to_excel('vanguard_etfs.xlsx')

# saving updated data to excel workbook and csv
updated_df_final.to_excel('vanguard_etfs_updated.xlsx')
updated_df_final.to_csv('vanguard_etfs_updated.csv')
```
